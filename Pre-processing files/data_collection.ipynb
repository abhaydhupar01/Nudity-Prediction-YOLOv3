```python
import urllib.request
import re
```

```python
with open(r'C:\Users\abhay\OneDrive\Desktop\Nudity Prediction\data collection txts\Abhay_boobs.txt','r') as f:
  i = 0  
  for line in f.readlines():
    image_url = line
    if not re.search('.jpg$', line):
        continue
    i+=1
    # Download the image and save it to a local file
    try:
        urllib.request.urlretrieve(image_url, f"data2/image{i}.jpg")
    except:
        continue
    print(f"Image{i} downloaded successfully!")
```
